```python
dang_ki ={"Minh","An","Hưng","Lan","Bình","Hùng"}
check_in ={"An","Hưng","Lan"}
```

```python
#In ra danh sách các bạn chưa check-in
chua_check_in = dang_ki - check_in
print(chua_check_in)
```

```python
#Tính toán tổng số lượng các bạn đã đăng ký và đã check-in.
tong = len(dang_ki) + len(check_in)
print("Tổng số lượng các bạn đã đăng ký và đã check-in : ",tong)
```

```python
#Sắp xếp danh sách các bạn đã đăng ký theo thứ tự từ điển.
print(sorted(dang_ki))
```
