```python
docker_compose = {
    "version": "3.8",
    "services": "app",
    "image": "python:3.10-slim",
    "command": "python app.py",
    "volumes": "./app:/app",
    "restart": "always",
    "ports": "5000:5000",
    "depends_on": "db"

}
```

```python
#In ra
print(docker_compose)
```

```python
#Sửa version thành '3.10', in kết quả.
docker_compose["version"]="3.10"
print("Sau khi sửa : ",docker_compose)
```

```python
#Lấy giá trị của key 'image' và in ra màn hình.
print(docker_compose.get("image"))
```

```python
#Thêm key 'environment' với giá trị ['PYTHONUNBUFFERED=1'], in kết quả.
docker_compose.update({"environment":['PYTHONUNBUFFERED=1']})
print(docker_compose)
```

```python
#Kiểm tra xem dictionary có chứa key 'volumes' hay không?
if "volumes" in docker_compose:
    print("Dictionary có chứa key 'volumes'")
else:
    print("Dictionary không chứa key 'volumes'")
```

```python
#Xóa key 'depends_on' khỏi dictionary, in kết quả.
docker_compose.pop("depends_on")
print(docker_compose)
```
