```python
import json

# Nhập từ bàn phím 1 dictionary
dict_input = json.loads(input("Nhập dictionary (dùng định dạng {'key': 'value', ...}): "))
```

```python
# Tạo dictionary mới bằng cách hoán đổi key và value
dict_swapped ={}
```

```python
for key, value in dict_input.items():
  if value in dict_swapped:
   #Nếu đã tồn tại làm key thì in ra NONE
   print(f"None")
   break
  dict_swapped[value] = key
else:
    print("Dict mới là :",dict_swapped)
```
