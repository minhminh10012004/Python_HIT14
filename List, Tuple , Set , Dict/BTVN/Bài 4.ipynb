```python
import random
password_list = ["CNTT", "KHMT", "KTPM", "HTTT", "DAPT"]
N = int(input("Nhập số lượng tài khoản (1<= N <= 100): "))
```

```python
accounts ={}
```

```python
for i in range(N):
  student_id = 2023600001 + i #MSV
  password = random.choice(password_list) + str(student_id)
  accounts[f"Account{i+1}"]={
      "Username": student_id,
      "Password": password
  }
```

```python
for account,detail in accounts.items():
  print(f"{account}: {detail}")
```
