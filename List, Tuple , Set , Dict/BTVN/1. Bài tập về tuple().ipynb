```python
#Nhập tuple A , B
A = tuple(map(int,input("Nhập các số nguyên ").split()))
B = tuple(input("Nhập các xâu kí tự ").split())
```

```python
#Hãy đếm số lượng phần tử trong A mà có giá trị lớn hơn trung bình cộng của tất cả phần tử
tb = sum(A)/len(A)
so_ptu_lon_hon_tb = len([x for x in A if x > tb])
print(f"Số lượng phần tử trong A mà có giá trị lớn hơn trung bình cộng của tất cả phần tử là {so_ptu_lon_hon_tb}")
```

```python
#Chia tuple A thành hai tuple con: một chứa các số chẵn và một chứa các số lẻ.
so_chan = tuple(x for x in A if x % 2 == 0)
so_le = tuple(x for x in A if x % 2 != 0)
print("Tuple số chẵn:",so_chan)
print("Tuple số lẻ:",so_le)
```

```python
#Nhập vào bàn phím ký tự k, đếm số lần xuất hiện của k trong B
k = input("Nhập kí tự k: ")
so_lan_xh = sum(i.count(k) for i in B)
print(f"Số lần xuất hiện của k trong B là {so_lan_xh}")
```

```python
#Tìm các phần tử trong tuple B có độ dài lớn hơn hoặc bằng 5 ký tự.
string = tuple(s for s in B if len(s) >= 5)
print("Các phần tử trong B có độ dài >= 5 là : ",string)
```

```python
#Kết hợp tuple A và tuple B thành một tuple mới C, trong đó các phần tử từ A và B được ghép cặp thành tuple con
C = tuple(zip(A,B))
print("Tuple mới C: ",C)
```
