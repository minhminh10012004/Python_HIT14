```python
#Sử dụng để lưu trữ các gtri dữ liệu trong các cặp key ; value
```

```python
#Ví dụ
sinhVien = {
    "hoVaTen" : "Đỗ Thị Minh",
    "maLop" : "A3",
    "diemTB" : 9.5
}
print(sinhVien)
print(sinhVien["hoVaTen"])
```

```python
#Sử dụng get để lấy giá trị
print(sinhVien.get("maLop"))
```

```python
#Thay đổi giá trị
#Cách 1
sinhVien["maLop"] = "A5"
print(sinhVien)
#Cách 2
sinhVien.update({"maLop":"A2","diemTB" : 10})
print(sinhVien)
```

```python
#Thêm các mục mới
sinhVien["gioiTinh"] = "Nữ"
print(sinhVien)
#Cách 2
sinhVien.update({"queQuan":"Hà Nội"})
print(sinhVien)
```

```python
#Xóa đi các mục
#Cách 1 : pop
sinhVien.pop("gioiTinh")
print(sinhVien)
#Cách 2 : del
del sinhVien["queQuan"]
print(sinhVien)
#Cách 3: popitem xóa mục key value cuối cùng
sinhVien.popitem()
print(sinhVien)
```

```python
#Xóa toàn bộ các mục
#sinhVien.clear()
#print(sinhVien)
```

```python
#Duyệt các gtri
for x in sinhVien.values():
    print(x)


```

```python
#Duyệt các khóa:
for x in sinhVien.keys():
    print(x)
```

```python
traiCay = {
    "Số 1":"Táo",
    "Số 2":"Cam",
    "Số 3":"Dưa Hấu"
}
#In tất cả các tên khóa trong từ điển , từng cái một ( gần giống với duyệt khóa)
for x in traiCay:
    print(x)
```

```python
#Duyệt các cặp khóa - gtri
for x,y in traiCay.items():
    print(x,y)
```
