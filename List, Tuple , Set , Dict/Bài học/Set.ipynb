```python
#Set không có thứ tự , gtri không được trùng nhau , không lặp chỉ mục , các mục set không thể thay đổi nhưng có thể xóa các mục và thêm mới
```

```python
#Tạo mới
monHoc ={"Toán","Lý","Hóa"}
print(monHoc)
```

```python
#Duyệt các phần tử trong set
for x in monHoc:
    print(x)
```

```python
#Thêm mới
#Cách 1 : Dùng add
monHoc.add("Sinh")
print(monHoc)

#Cách 2 : Dùng update
hocThem ={"Anh","Văn"}
monHoc.update(hocThem)
print(monHoc)
```

```python
#Thêm list vào SET
hocPhuDao =["Võ thuật","Múa","Múa"] #=>Thêm vào set thì chỉ có 1 Múa (Set không cho gtri trùng nhau)
print(hocPhuDao)
monHoc.update(hocPhuDao)
print(monHoc)
```

```python
#Xóa
#Cách 1 : Dùng remove (Nếu phần tử ko tồn tại sẽ phát sinh lỗi)
#monHoc.remove("Múa")
#print(monHoc)

#Cách 2 : Dùng discard (Nếu ptu ko tồn tại sẽ không phát sinh lỗi)
monHoc.discard("Anh")
print(monHoc)
```

```python
#Xóa phần tử đầu tiên (pop)
monHoc.pop()
print(monHoc)
```

```python
#Xóa sạch (clear)
monHoc.clear()
print(monHoc)
```

```python
#Xóa bỏ tập hợp bằng từ khóa (del)
del monHoc
print(monHoc)
```
