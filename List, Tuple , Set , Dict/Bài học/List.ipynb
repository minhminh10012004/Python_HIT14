```python
studentList = ["An","Bình","Minh"]
#List có thứ tự , vị trí các phần tử được đánh dấu từ 0 , từ trái qua phải

print(studentList[2])
print(studentList[0])
```

```python
print(studentList)
print(studentList[:])
```

```python
#studentList[x:y] => Lấy ra [x:y)
print(studentList[1:2])
print(studentList[0:2])
```

```python
#Thêm phần tử vào cuối List
studentList.append("Hùng")
print(studentList)

#Cách khác
studentList[len(studentList):] = ["Thành"]
print(studentList)
```

```python
#Chèn phần tử vào list
studentList.insert(1,"Hưng")
print(studentList)
```

```python
#Số lượng phần tử trong list
print(len(studentList))
```

```python
#Đếm số lượng phần tử thỏa điều kiện
print("Đếm Hưng: ",studentList.count("Hưng"))
print("Đếm Thành: ",studentList.count("Thành"))
```

```python
#Kiểm tra phần tử có trong list: in

#Xóa phần tử ra khỏi list
if "Thành" in studentList:
  studentList.remove("Thành")
  print(studentList)

#Xóa phần tử ra khỏi list bằng vị trí
studentList.pop(0) #=> xóa An
print(studentList)
```

```python
#Đảo ngược list
studentList.reverse()
print(studentList)
```

```python
#Sắp xếp lại
studentList.sort()
print(studentList)
```

```python
numbers = [5,6,3,8,4]
numbers.sort()
print(numbers)
```

```python
#Sắp xếp ngược
studentList.sort(reverse=True)
print(studentList)

numbers.sort(reverse=True)
print(numbers)
```

```python
#Xóa sạch dữ liệu trong list
studentList.clear()
print(studentList)
```
