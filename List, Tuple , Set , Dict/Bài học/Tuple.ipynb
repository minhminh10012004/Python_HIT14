```python
#Giá trị của Tuple là không thể thay
gioiTinh = ("Nam", "Nữ")
traiCay = ("Táo", "Chuối", "Táo","Cam")
```

```python
print(gioiTinh[0])
print(traiCay[0:2])
```

```python
for x in traiCay:
    print(x)
```

```python
#Cộng 2 tuple
y =(1,2,3) + (4,5,6)
print(y)
```

```python
#Nhân 2 tuple
z = y * 2
print(z)
```

```python
#Kiểm tra dữ liệu có nằm trong tuple không
print("Táo" in traiCay)
print("Bom" in traiCay)
```

```python
#Lấy độ dài của Tuple
print(len(traiCay))
print(len(gioiTinh))
```

```python
#Đếm số lượng xuất hiện
print(traiCay.count("Táo"))
```

```python
#Tìm min , max
print(min(gioiTinh)) #=>theo anpha
print(max(traiCay)) #=>theo anpha

#Tính sum (chỉ dùng được cho Tuple là số )
numbers = (1,2,3,4,5)
print(sum(numbers))
print(min(numbers))
print(max(numbers))
```

```python
listTC = sorted(traiCay) => chuyển thành list sếp theo anpha
print(listTC)
```
